```python
import numpy as np
from matplotlib import pyplot as plt

# Создаем массив x из 51 числа от -3 до 3 включительно
x = np.linspace(-3, 3, 51)

# Определяем функции для массивов y1, y2, y3, y4
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

# Создаём объект Figure с названием fig и массив объектов Axes под названием ax
fig, ax = plt.subplots(2, 2)

# Устанавливаем границы по оси x для левого верхнего графика
ax[0, 0].set_xlim(-5, 5)

# Рисуем графики
ax[0, 0].plot(x, y1)
ax[1, 0].plot(x, y2)
ax[0, 1].plot(x, y3)
ax[1, 1].plot(x, y4)

# Даём названия графикам
ax[0, 0].set_title('График y1')
ax[1, 0].set_title('График y2')
ax[0, 1].set_title('График y3')
ax[1, 1].set_title('График y4')

# Устанавливаем размеры фигуры
plt.figure(figsize=(8, 6))

# Устанавливаем вертикальные и горизонтальные зазоры между графиками
plt.tight_layout(pad=0.3)

# Показываем график
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('creditcard.csv')
counts = df['Class'].value_counts()
log_counts = np.log(counts)

ax = counts.plot(kind='bar', figsize=(10, 5), style='fivethirtyeight')
ax.set_title('Количество наблюдений по классу')
plt.show()

ax = log_counts.plot(kind='bar', figsize=(10, 5), style='fivethirtyeight', logy=True)
ax.set_title('Логарифмическое количество наблюдений по классу')
plt.show()

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 6))

# Гистограмма обычных транзакций
ax1.hist(df[df['Class'] == 0]['V1'], bins=20, density=True, alpha=0.5, color='grey')
ax1.set_xlabel('V1')
ax1.set_ylabel('Плотность распределения')
ax1.legend(['Class 0'])

# Гистограмма мошеннических
ax2.hist(df[df['Class'] == 1]['V1'], bins=20, density=True, alpha=0.5, color='red')
ax2.set_xlabel('V1')
ax2.set_ylabel('Плотность распределения')
ax2.legend(['Class 1'])

ax1.set_title('Гистограмма обычных транзакций')
ax2.set_title('Гистограмма мошеннических транзакций')
plt.show()
```

```python
import numpy as np

# Создание одномерного массива а от 0 до 12
a = np.arange(0, 12)

# Изменение формы массива а, чтобы получить матрицу A 4x3
A = a.reshape(4, 3)

# Транспонирование матрицы A
At = A.T

# Умножение матрицы A на матрицу At
B = np.dot(A, At)

# Размер матрицы B
print("Размер матрицы B:", B.shape)

# Попытка вычислить обратную матрицу для матрицы B
try:
    np.linalg.inv(B)
    print("Обратная матрица для матрицы B существует.")
except np.linalg.LinAlgError:
    print("Обратной матрицы для матрицы B не существует.")
```

После выполнения этого кода вы получите размер матрицы B, который будет (4, 4). Однако, так как матрица B получается путем умножения матрицы A саму на себя транспонированную (A * At), матрица B будет вырожденной (сингулярной) и не будет иметь обратной матрицы.

```python
import numpy as np

# Инициализация генератора случайных чисел
np.random.seed(42)

# Создание одномерного массива c из 16 случайных целых чисел от 0 до 15
c = np.random.randint(0, 15, 16)

print("Массив c:")
print(c)
```

```python
import numpy as np

# Изменение формы массива c, чтобы получить квадратную матрицу C
C = c.reshape(4, 4)

# Получение матрицы D
D = C * 10 + B

# Вычисление определителя, ранга и обратной матрицы D
det_D = np.linalg.det(D)
rank_D = np.linalg.matrix_rank(D)

try:
    D_inv = np.linalg.inv(D)
    print("Матрица D:")
    print(D)
    print("Определитель матрицы D:", det_D)
    print("Ранг матрицы D:", rank_D)
    print("Обратная матрица D:")
    print(D_inv)
except np.linalg.LinAlgError:
    print("Обратная матрица для матрицы D не существует, так как матрица D вырожденная.")
```

```python
import numpy as np

# Приравняем отрицательные числа в матрице D_inv к 0, а положительные к 1
D_inv_modified = np.where(D_inv < 0, 0, 1)

# Создадим матрицу E с помощью условий из матриц D_inv, B и C
E = np.where(D_inv_modified, B, C)

print("Матрица E:")
print(E)
```
