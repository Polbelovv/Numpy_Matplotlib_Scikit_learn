```python
import sympy as sp

# Определим переменную и функцию
x = sp.symbols('x')
f = x * sp.log(x)

# Найдем производную
f_prime = sp.diff(f, x)

# Решим уравнение f'(x) = 0 для нахождения критических точек
critical_points = sp.solve(f_prime, x)

# Определим интервалы возрастания и убывания
intervals = [(0, critical_points[0]), (critical_points[0], sp.oo)]
signs = [sp.sign(f_prime.subs(x, point)) for point in [0.1, 1]]

# Вывод результатов
print("Критические точки:", critical_points)
print("Интервалы убывания:", intervals[0] if signs[0] < 0 else "Нет")
print("Интервалы возрастания:", intervals[1] if signs[1] > 0 else "Нет")
```

```python
import sympy as sp

# Определяем переменную и функцию
x = sp.symbols('x')
y = 1 / (1 - x**2)

# Находим производную функции
y_prime = sp.diff(y, x)

# Находим критические точки (где производная равна нулю или не существует)
critical_points = sp.solveset(y_prime, x, domain=sp.Reals)

# Определяем интервалы возрастания и убывания
increasing_intervals = sp.solve_univariate_inequality(y_prime > 0, x, relational=False)
decreasing_intervals = sp.solve_univariate_inequality(y_prime < 0, x, relational=False)

print(f"Производная функции: {y_prime}")
print(f"Критические точки: {critical_points}")
print(f"Интервалы возрастания: {increasing_intervals}")
print(f"Интервалы убывания: {decreasing_intervals}")
```

экстремумы функий

```python
import sympy as sp

x = sp.symbols('x')
f = x**3 - 3*x + 1
f_prime = sp.diff(f, x)
f_double_prime = sp.diff(f_prime, x)

# Решение уравнения производной
critical_points = sp.solve(f_prime, x)

# Проверка экстремумов
extrema = []
for point in critical_points:
    second_derivative = f_double_prime.subs(x, point)
    if second_derivative > 0:
        extrema.append((point, f.subs(x, point), 'Максимум'))
    elif second_derivative < 0:
        extrema.append((point, f.subs(x, point), 'Минимум'))

for ext in extrema:
    print(f"Значение функции на экстремуме: {ext[1]}, экстремум: {ext[2]} в точке x = {ext[0]}")
```

```python
import sympy as sp

# Определяем переменную
x = sp.symbols('x')

# Определяем функцию
y = sp.exp(x**2 - 4*x + 5)

# Находим первую производную
y_prime = sp.diff(y, x)

# Находим критические точки
critical_points = sp.solve(y_prime, x)

# Находим вторую производную
y_double_prime = sp.diff(y_prime, x)

# Проверяем характер критических точек
extrema = []
for point in critical_points:
    second_derivative = y_double_prime.subs(x, point)
    if second_derivative > 0:
        extrema.append((point, 'Minimum'))
    elif second_derivative < 0:
        extrema.append((point, 'Maximum'))
    else:
        extrema.append((point, 'Saddle point'))

print("Критические точки и их характер:")
for point, nature in extrema:
    print(f"x = {point}, {nature}")
```

```python
import sympy as sp

# Определяем переменную и функцию
x = sp.symbols('x')
y = x - sp.atan(x)

# Производная
y_prime = sp.diff(y, x)

# Решаем уравнение y' = 0
critical_points = sp.solve(y_prime, x)

# Вторая производная
y_double_prime = sp.diff(y_prime, x)

# Анализируем вторую производную в критических точках
extremum_points = []
for point in critical_points:
    second_derivative_value = y_double_prime.subs(x, point)
    if second_derivative_value > 0:
        extremum_type = 'minimum'
    elif second_derivative_value < 0:
        extremum_type = 'maximum'
    else:
        extremum_type = 'неубедительный'
    extremum_points.append((point, extremum_type))

# Выводим результаты
print(f"Критические точки: {critical_points}")
print(f"Вторая производная: {y_double_prime}")
print("Экстремумы:")
for point, extremum_type in extremum_points:
    print(f"x = {point}: {extremum_type}")
```

Найти интервалы выпуклости и точки перегиба функций:

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-x**2)

def df(x):
    return -2*x*np.exp(-x**2)

def d2f(x):
    return (4*x**2 - 2)*np.exp(-x**2)

# Находим точки, где вторая производная равна нулю
x_points = np.roots([4, 0, -2])

# Сортируем точки для удобства
x_points.sort()

# Создаем интервалы
intervals = [(-np.inf, x_points[0]), (x_points[0], x_points[1]), (x_points[1], np.inf)]

# Определяем выпуклость на каждом интервале
for interval in intervals:
    x_test = (interval[0] + interval[1]) / 2  # Берем тестовую точку в середине интервала
    if d2f(x_test) > 0:
        print(f"Функция выпукла вверх на интервале {interval}")
    else:
        print(f"Функция выпукла вниз на интервале {interval}")

# Выводим точки перегиба
print(f"Точки перегиба: x = {x_points[0]}, x = {x_points[1]}")

# Строим график для визуализации
x = np.linspace(-3, 3, 500)
y = f(x)
plt.plot(x, y)
plt.plot(x_points, f(x_points), 'ro', label='Точки перегиба')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График функции f(x) = e^(-x^2)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем функцию
def f(x):
  return np.cos(x)

# Определяем вторую производную функции
def f_2prime(x):
  return -np.cos(x)

# Создаем массив значений x
x = np.linspace(-2*np.pi, 2*np.pi, 1000)

# Вычисляем значения y и второй производной
y = f(x)
y_2prime = f_2prime(x)

# Находим точки перегиба
inflection_points = x[np.where(np.diff(np.sign(y_2prime)))]

# Выводим результаты
print("Точки перегиба:")
for point in inflection_points:
  print(f"x = {point:.2f}")

# Строим график функции, точек перегиба и областей выпуклости/вогнутости
plt.plot(x, y, label='y = cos(x)')
plt.plot(inflection_points, f(inflection_points), 'ro', label='Точки перегиба')
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции y = cos(x), точки перегиба и выпуклость')

# Отмечаем области выпуклости и вогнутости
for i in range(len(inflection_points) - 1):
  if f_2prime((inflection_points[i] + inflection_points[i+1])/2) > 0:
    plt.fill_between(x[np.logical_and(x >= inflection_points[i], x <= inflection_points[i+1])], 
                    y[np.logical_and(x >= inflection_points[i], x <= inflection_points[i+1])], 
                    color='green', alpha=0.3)
  else:
    plt.fill_between(x[np.logical_and(x >= inflection_points[i], x <= inflection_points[i+1])], 
                    y[np.logical_and(x >= inflection_points[i], x <= inflection_points[i+1])], 
                    color='red', alpha=0.3)

plt.legend()
plt.grid(True)
plt.show()

print("\nИнтервалы выпуклости и вогнутости:")
for i in range(len(inflection_points) - 1):
  if f_2prime((inflection_points[i] + inflection_points[i+1])/2) > 0:
    print(f"Выпуклая на интервале ({inflection_points[i]:.2f}, {inflection_points[i+1]:.2f})")
  else:
    print(f"Вогнутая на интервале ({inflection_points[i]:.2f}, {inflection_points[i+1]:.2f})")
```

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# Определяем символьную переменную
x = sp.Symbol('x')

# Определяем функцию
y = x**5 - 10*x**2 + 7*x

# Находим первую производную
dy = sp.diff(y, x)

# Находим вторую производную
d2y = sp.diff(dy, x)

print("Вторая производная:", d2y)

# Находим корни второй производной
roots = sp.solve(d2y, x)

# Отбираем только реальные корни
real_roots = [root.evalf() for root in roots if root.is_real]

print("Корни второй производной:", real_roots)

# Определяем интервалы выпуклости
test_points = [-np.inf] + real_roots + [np.inf]
convexity = []

for i in range(len(test_points) - 1):
    mid = (test_points[i] + test_points[i+1]) / 2
    sign = d2y.subs(x, mid).evalf()
    if sign > 0:
        convexity.append((test_points[i], test_points[i+1], "выпуклая вверх"))
    elif sign < 0:
        convexity.append((test_points[i], test_points[i+1], "выпуклая вниз"))

print("Интервалы выпуклости:")
for interval in convexity:
    print(f"({interval[0]}, {interval[1]}): {interval[2]}")

# Находим точки перегиба
inflection_points = []
for root in real_roots:
    y_value = y.subs(x, root).evalf()
    inflection_points.append((float(root), float(y_value)))

print("Точки перегиба:", inflection_points)

# Визуализация
x_vals = np.linspace(-3, 3, 1000)
y_vals = [y.subs(x, val).evalf() for val in x_vals]

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals)
plt.title('График функции')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
for point in inflection_points:
    plt.plot(point[0], point[1], 'ro')
plt.show()
```

Провести полное исследование и построить графики функций

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# Определение функции
def func(x):
    return np.log(1 - x**2)

# Границы интервала для построения графика
x_min = -1.2
x_max = 1.2

# Шаг для построения графика
x_step = 0.01

# Генерация массива значений x
x_values = np.arange(x_min, x_max + x_step, x_step)

# Вычисление соответствующих значений y
y_values = func(x_values)

# Построение графика функции
plt.plot(x_values, y_values, label='y = ln(1 - x^2)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции y = ln(1 - x^2)')
plt.grid(True)
plt.legend()
plt.show()

# Нахождение корней функции
roots = fsolve(func, 0)
print(f'Корни функции: {roots}')

# Нахождение экстремумов функции
derivative = lambda x: -2 * x / (1 - x**2)
extremum_points = fsolve(derivative, 0)
extremum_values = func(extremum_points)
print(f'Экстремумы функции: {extremum_points} с значениями {extremum_values}')

# Исследование знака производной
signs = np.sign(derivative(x_values))
print(f'Знаки производной на интервале [{x_min}, {x_max}]: {signs}')
```
